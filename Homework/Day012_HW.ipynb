### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text)
nslist = soup.find(class_ = 'part_list_2').find_all('h3')
for ns in nslist:
    print(ns.find('span').text, ns.find('em').text, ns.find('a').text)
```
