### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.ptt.cc/bbs/NBA/index6504.html'
r = requests.get(url, 'html5lib')
soup = BeautifulSoup(r.text)
dates = soup.find_all(class_='date') #因為第一頁有置頂貼文，最新貼文必須視時間而定
ds = []
for d in dates:
    ds.append(d.text)
    
from collections import defaultdict
x = defaultdict(list)
for k,va in [(v,i) for i,v in enumerate(ds)]:
    x[k].append(va)
print(x) #取出所有日期的日子所在的位置
print('最新貼文的位置為：' + str(max(x['12/23']))) #找出12/23(今天)最新貼文位置為3

latestpost = soup.find_all(class_='title')[3].find('a') #取出該位置的貼文資訊
lpurl = 'https://www.ptt.cc' + latestpost['href']
lpweb = BeautifulSoup(requests.get(lpurl, 'html5lib').text)
lpinfo = lpweb.find_all(class_='article-meta-value')

print('最新一篇貼文資訊如下：')
print('Author: '+lpinfo[0].text)
print('Title: '+lpinfo[2].text)
print('Time: '+lpinfo[3].text)
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
p1_posts = list(soup.find_all(class_='title'))
for i in range(len(p1_posts)):
    u = 'https://www.ptt.cc'+p1_posts[i].find('a')['href']
    web = BeautifulSoup(requests.get(u, 'html5lib').text)
    info = web.find_all(class_='article-meta-value')
    try:
        print('\nAuthor: '+info[0].text)
        print('Title: '+info[2].text)
        print('Time: '+info[3].text)
    except:
        continue
#投票的貼文取不出來?
```

### ③ 試著爬爬看其他版的文章
