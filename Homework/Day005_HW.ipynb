```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests as req
import json
r2 = req.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(r2.text) #dataç‚ºlist 
print('ç¸½å…±å›å‚³'+str(len(data))+'ç­†è³‡æ–™ã€‚')
print('\næ¯ç­†è³‡æ–™æ¬„ä½å¦‚ä¸‹ï¼š')
for key in data[0].keys(): #data[0]ç‚ºdict
    print(key)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

print("æ¯ç­†è³‡æ–™æ¨™é¡Œå¦‚ä¸‹ï¼š")
for d in data:
    print(d['title'])
print("\næ¯ç­†è³‡æ–™è²¼æ–‡æ™‚é–“å¦‚ä¸‹ï¼š")
for d in data:
    print(d['createdAt'])
print("\næ¯ç­†è³‡æ–™ç•™è¨€äººæ•¸å¦‚ä¸‹ï¼š")
for d in data:
    print(d['commentCount'])
print("\næ¯ç­†è³‡æ–™æŒ‰è®šäººæ•¸å¦‚ä¸‹ï¼š")
for d in data:
    print(d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

import requests as req
import json

a = req.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true') #ç†±é–€æ–‡ç« 
data_a = json.loads(a.text)
b = req.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false') #éç†±é–€æ–‡ç« 
data_b = json.loads(b.text)

commentCount_a = [] 
likeCount_a = []
commentCount_b = []
likeCount_b = []

for d1 in data_a:
    commentCount_a.append(d1['commentCount']) #æŠŠç•™è¨€æ•¸èˆ‡è®šæ•¸å­˜å…¥åˆ—è¡¨
    likeCount_a.append(d1['likeCount'])
for d2 in data_b:
    commentCount_b.append(d2['commentCount'])
    likeCount_b.append(d2['likeCount'])
    
print(sum(commentCount_a)/len(commentCount_a)) #ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€æ•¸
print(sum(likeCount_a)/len(likeCount_a)) #ç†±é–€æ–‡ç« å¹³å‡è®šæ•¸
print(sum(commentCount_b)/len(commentCount_b)) #éç†±é–€æ–‡ç« å¹³å‡ç•™è¨€æ•¸
print(sum(likeCount_b)/len(likeCount_b)) #éç†±é–€æ–‡ç« å¹³å‡è®šæ•¸
```
