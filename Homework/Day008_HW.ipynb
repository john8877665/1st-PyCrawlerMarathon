### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'
r = requests.get(url)

r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')

print(type(r.text)) #是str不能直接取中間的資料
soup = BeautifulSoup(r.text, "html5lib")
print(soup)
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')
print("因為變成BeautifulSoup資料後在python中較易處理，可以輕易取出資料。")
print(type(soup)) #變成bs4.BeautifulSoup的資料

print("\nFor example: 取出看板與校園列表所有主題"+"\n")
for i in range(len(soup.find_all('a',class_="ForumEntry_link_3tNC45 ForumEntry_unRead_1ptlOL"))):
    print(soup.find_all('a',class_="ForumEntry_link_3tNC45 ForumEntry_unRead_1ptlOL")[i].text)
```

### 2. 知乎： https://www.zhihu.com/explore

```python
url2 = 'https://www.zhihu.com/explore'
r2 = requests.get(url2)
r2.encoding = 'utf-8'

print(r2.text[0:600])
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'

hds = {'user-agent': 'my-app/0.0.1'} #知乎要求 user-agent 的 request header
r2 = requests.get("https://www.zhihu.com/explore", headers=hds)
res2 = r2.text
print(res2)
```
